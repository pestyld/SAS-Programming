```sas
**********************************************************************;
* STUDENT QUESTION: Changing the title of by groups in proc print    *;
**********************************************************************;
* PROGRAM DESCRIPTION: Change the title of each group when using the *;
*                      by statement in proc print                    *;
**********************************************************************;
```

```sas
**************************************;
*The default report                  *;
* Notice the title is Origin='value' *;
**************************************;

* Need to sort the data by the by group *;
proc sort data=sashelp.cars
          out=cars;
    by Origin;
run;

* Print each group with a title *;
proc print data=cars;
    by Origin;
run;
```

```sas
***********************************;
*Changing the titles in each group ;
***********************************;
options nobyline; /*<---Need to do this to turn off defaults*/

title "#byval1 is the value";/*Uses the BY value in the title*/
title2 "#byvar1 is the column name"; /*Uses the BY column name*/
title3 "The #byvar1 is #byval1";/*Uses both*/

proc print data=cars;
    by Origin;
run;

title;
options byline; /*<---Return to the default setting*/
```
