# How can I see the number of rows in a table?

Shows how to count the number of rows in an entire table. Also how to count the number of non missing rows in a column. Eventually create an macro program on how to do it. 

## Counts total number of rows in a table

```sas
proc sql;
select count(*) as TotalRows
    from sashelp.cars;
quit;
```

## Count total number of nonmissing rows in a specific column

```sas
* Specify column to count *;
%let colName=Make;

title "Count total number of nonmissing rows in column: &colName";

proc sql;
select count(&colName) as TotalRows
    from sashelp.cars;
quit;

title;
```

## Create a macro program that does the same thing as above. Now you can easily type %len(table name) and be done

```sas
%MACRO len(table, col=*);
* table = table to count *;
* col = column to count. By default counts all rows in a table. If a column is specified, will count number of non missing rows*;

proc sql;
    select count(&col) as TotalRows
    from &table;
quit;

%mend;
```

```sas
* Count all rows in a table using the macro program *;
%len(sashelp.cars)
```

```sas
* Count number of non missing rows in a specific column *;
%len(sashelp.cars, col=Cylinders)
```
