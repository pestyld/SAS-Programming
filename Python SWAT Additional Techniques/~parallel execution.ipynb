```python
conn1 = swat.CAS('server.demo.sas.com', 30571, 'student', 'Metadata0', name='py04d04_1')
conn2 = swat.CAS('server.demo.sas.com', 30571, 'student', 'Metadata0', name='py04d04_2')
conn3 = swat.CAS('server.demo.sas.com', 30571, 'student', 'Metadata0', name='py04d04_3')
```

```python
loansTbl = {'name':'loans_raw', 'caslib':'casuser'}
appTbl = {'name':'appRatings', 'caslib':'casuser'}
custTbl = {'name':'customers_raw', 'caslib':'casuser'}
```

```python
for i in :
    print(i)
```

```python
%%time
connections = [conn1, conn2, conn3]
tables = [loansTbl, appTbl, custTbl]

for i in range(3):
    connections[i].invoke('distinct', table =tables[i])
    print('-------------------Executed invoke----------------------')
```

```python
ds1 = """
data casuser.test1;
    do i=1 to 50000000;
        cat = rand('table',.5,.5);
        num = rand('uniform',1,100000);
        output;
    end;
run;"""

ds2 = """
data casuser.test2;
    do i=1 to 1000000;
        cat = rand('table',.5,.5);
        num = rand('uniform',1,100000);
        output;
    end;
run;"""

ds3 = """
data casuser.test3;
    do i=1 to 100;
        cat = rand('table',.5,.5);
        num = rand('uniform',1,100000);
        output;
    end;
run;"""


#table={'name':'test1', 'groupBy':'cat'}
```

```python
conn1.invoke('runCode', code = ds1 )
conn1.tableInfo()
```

```python
conn2.invoke('runCode', code = ds2 )
conn2.tableInfo()
```

```python
conn3.invoke('runCode', code = ds3 )
conn3.tableInfo()
```

```python
conn1.invoke('summary', table='test1')
conn2.invoke('summary', table='test2')
conn3.invoke('summary', table='test3')

for resp, conn in swat.getnext([conn1, conn2, conn3]):
    print('value of response:', resp)
    for k, v in resp:
        print('-------------------')
        print('>>> RESULT', k, v)
```
